## Installation Guide for Jupyter Notebook

Throughout this course, Jupyter Notebook is used to perform Lab exercises.

[Jupyter Notebook](https://jupyter-notebook.readthedocs.io/en/stable/notebook.html) is an open-source, interactive web tool that one can use to create and share documents that contain live code, equations, visualizations, and text.

The name, Jupyter, comes from core programming languages that it supports: Julia, Python, and R. Jupyter is originally based on IPython kernel, which allows one to write your programs in Python.

To run Jupyter Notebook for the Lab exercises, you need to have installed Python environment. 

## 1. Installation
The easiest way to install Python and Jupyter Notebook at the same time is by using Anaconda [Anaconda](https://www.anaconda.com/products/individual) distibution (option 1). Alternatively, you can install Python first and then Jupyter Notebook (Option 2).

### Option 1: Anaconda3 (Recommended)
Anaconda distribution simplifies installation and package managment for Python.

- Go to Anaconda download page (https://docs.anaconda.com/anaconda/install/) and find the installer file that matches your system
- Open the file, start the installation wizard and follow the instruction
- After installation, open Jupyter Notebook by running the on the terminal `jupyter notebook`

    #### Virtual environments with Anaconda
    A virtual environment is a self-contained directory tree that contains a Python installation for a particular version of Python, plus a number of additional packages.

    It is a good idea to create and use a virtual environment that will contain important packages useful for the Lab exercises instead of doing system-wide installation.

    You can use the Anaconda Navigator (graphical interface) or conda (via the terminal):

   - Create a virtual environment 
       - `$ conda create --name [env-name]` e.g., you can use SEAI as env name
   - Activate a virtual environment
       - `$ source activate [env-name]` Linux or MacOS
       - `$ activate [env-name]` Windows
   - Deactivate a virtual environment
       - `$ source deactivate` Linux or MacOS
       - `$ deactivate` Windows

#### Install Packages
   You need the following Python packages for the exercises: ``` numpy scipy matplotlib scikit-learn pandas```

  To install them, you can use the Anaconda Navigator (graphical interface) or conda (via terminal):
        ```(base)$ conda install numpy scipy matplotlib scikit-learn pandas```

### Option 2: Python and Jupyter Notebook separately 
To install Python and then Jupyter Notebook separately follow the steps below.

   #### Install Python
   Check if Python3 is installed. On terminal, write the following commands:
   - Windows: ```python ```
   - Linux or MacOS: ``` $ python -V ```

   You should see as output the installed Python version number.
If not, install Python3 as described below.

       ###### On Windows
        - Go to Python download page (https://www.python.org/downloads/windows/) and find windows installer file that matches your system.
        - Open file, start the installation wizard and follow instructions. Take note of where python is installed

    ###### On Linux or MacOS 
     - Follow the steps described in this guideline:(https://wsvincent.com/install-python3-mac/)

#### Install Jupyter Notebook
You can follow installation steps here (https://jupyter.readthedocs.io/en/latest/install.html)
- Ensure latest pip is installed (pip comes with Python3)
    - ```pip3 install --upgrade pip```
- Install Jupyter Notebook
    - ``` pip3 install jupter```

- Open and run Jupyter Notebook
    - ```jupyter notebook```
    
    #### Virtual environments with Python
    It is a good idea to create the virtual environments in one place. For instance, on MacOS and Linux, a standard location is ```$ HOME/.virtualenvs```

    ```$ mkdir ~/.virtualenvs```

    A tool for creating isolated Python environment is venv (Python 3)

        - Create a virtual environment
            - `$ cd ~/.virtualenvs`
            - `$ python3 -m venv [env-name]`(e.g., you can use SEAI as env name)
        - Activate a virtual environment
            - `$ source $HOME/.virtualenvs/[env-name]/bin/activate`(Linux or MacOS)
            - `$ [env-name]\Scripts\activate.bat` (Windows)

       - Deactivate a virtual environment(env-name)
            - `$ deactivate`


## 2. Working with Jupyter Notebooks

### Anaconda
- Go to start menu and click on Anaconda
- Launch Jupyter notebook(mostly second panel) from the Anaconda Navigator
- A browser window should open up with your directory listing.
- You can create your own folder and navigate to that folder to save the scripts.
- Click on New at the top right end and choose “Python 3” from drop downmenu.
- You can start writing your code in the grey box!
- Try executing the code with the “Run” button or Alt-Enter

### From Terminal
To open an existing notebook file(env-name)
    `$ jupyter notebook filename.ipynb`
